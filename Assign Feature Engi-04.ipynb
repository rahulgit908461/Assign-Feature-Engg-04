```python
# suppose we are taking one data from seaborn 
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
# here in this tips data total columns are 7 and total rows is 244 then we are going to use nominal encoding to
# perform the the categorical data and check how many new columns be created
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
df.columns
```

```python
df1=df[['sex','day'   , 'time' , 'size']]
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df1[['sex','day'   , 'time' , 'size']]).toarray()
```

```python
encoded=encoder.fit_transform(df1[['sex','day'   , 'time' , 'size']])
```

```python
pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
pd.concat([df1,encoded_df],axis=1)
```

```python
df2=pd.concat([df1,encoded_df],axis=1)
```

```python
# here we can see as per unique data columns are changed but in rows no any changing  so if in two column are catgorical
# if both columns is unique then 2 column are present after using nominal encoding to transform the catgorical data.
```
